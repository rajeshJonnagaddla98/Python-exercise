```python
import requests
from bs4 import BeautifulSoup

r = requests.get("https://www.flipkart.com/search?q=washing+machine&sid=j9e%2Cabm%2C8qx&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=washing+machine%7CWashing+Machines&requestId=c1d6b0c8-6612-4ac0-bbd8-7422e574a5e9&p%5B%5D=facets.price_range.from%3D25000&p%5B%5D=facets.price_range.to%3DMax")
c = r.content
print(c)
```

```python
soup = BeautifulSoup(c,"html.parser")
```

```python
soup
```

```python
all=soup.find_all("div",{"class":"tUxRFH"})
all
```

```python
for item in all:   #to display all price value in the website
    print(item.find_all("div",{"class":"KzDlHZ"})[0].text)
    print(item.find_all("div",{"class":"Nx9bqj _4b5DiR"})[0].text)
    print(item.find_all("div",{"class":"XQDdHH"})[0].text)
    print(item.find_all("div",{"class":"yKfJKb row"})[0].text)
```

```python
l=[]
for item in all:   #to display all price value in the website
    d={}
    d["washing_machine_name"]=item.find_all("div",{"class":"KzDlHZ"})[0].text
    d["washing_machine_price"]=item.find_all("div",{"class":"Nx9bqj _4b5DiR"})[0].text.replace("","Rs.")
    d["washing_machine_ratings"]=item.find_all("div",{"class":"XQDdHH"})[0].text
    d["washing_machine_description"]=item.find_all("div",{"class":"yKfJKb row"})[0].text
    l.append(d)
```

```python
import pandas
df=pandas.DataFrame(l)
```
